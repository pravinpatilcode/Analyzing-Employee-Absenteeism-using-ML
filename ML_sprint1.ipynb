```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

def feature_scale(df, scale='standard'):
    numerics = ['int16', 'int32', 'float16', 'float32', 'float64']
    new_data = df.select_dtypes(include=numerics)  # Select only numeric columns

    # Correct the comparison to use '=='
    if scale == 'minmax':
        scaler = MinMaxScaler()  # Use Min-Max Scaler if 'minmax' is chosen
    else:
        scaler = StandardScaler()  # Use Standard Scaler as default

    # Fit and transform the data
    scaled_data = scaler.fit_transform(new_data)

    # Return the scaled data as a DataFrame, preserving column names
    df_scaled = pd.DataFrame(scaled_data, columns=new_data.columns)

    return df_scaled
```

```python
# Assuming 'df' is your DataFrame
scaled_df = feature_scale(df, scale='minmax')
print(scaled_df.head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming 'cat_loan_df' is your DataFrame with categorical columns

# Select the categorical columns
cat_loan_df = df.select_dtypes('category')

# Get the categorical columns
categorical_cols = cat_loan_df.columns

# Compute number of rows for plot
num_cols = 3
num_rows = int(np.ceil(len(categorical_cols) / num_cols))

# Setting canvas for plotting
fig, axs = plt.subplots(num_rows, num_cols, tight_layout=True, figsize=(20, 5 * num_rows))

# Flatten axs array in case it's multi-dimensional
axs = axs.flatten()

# Plotting the categorical columns
for col, ax in zip(categorical_cols, axs[:len(categorical_cols)]):
    stats = cat_loan_df[col].value_counts()
    values = list(stats)
    names = list(map(str, stats.index))

    ax.bar(names, values)
    ax.set_title(col)
    ax.set_xticklabels(names, rotation=45, ha="right")

# Hide unused subplots
for ax in axs[len(categorical_cols):]:
    ax.set_visible(False)

plt.show()
```
